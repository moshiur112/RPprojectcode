```python
import pandas as pd
import os
import re
from matplotlib import pyplot as plt
from scipy.stats import pearsonr
```

```python
addg4 = 'imagestoaddg4sorted.txt'
addg3 = 'imagestoaddg3sorted.txt'
addg2 = 'imagestoaddg2sorted.txt'
addg1 = 'imagestoaddg1sorted.txt'
```

```python
all_images = []
with open(addg1) as f:
    
    lines = f.readlines()
    imageRE = 'IM_[0-9a-zA-Z]{11,15}$'
    for line in lines:
        graphicID = re.findall(imageRE, line)
        all_images.append(graphicID[0])

with open(addg2) as f:
    lines = f.readlines()
    imageRE = 'IM_[0-9a-zA-Z]{11,15}$'
    for line in lines:
        graphicID = re.findall(imageRE, line)
        all_images.append(graphicID[0])   
with open(addg3) as f:
    lines = f.readlines()
    imageRE = 'IM_[0-9a-zA-Z]{11,15}$'
    for line in lines:
        graphicID = re.findall(imageRE, line)
        all_images.append(graphicID[0])
with open(addg4) as f:
    lines = f.readlines()
    imageRE = 'IM_[0-9a-zA-Z]{11,15}$'
    for line in lines:
        graphicID = re.findall(imageRE, line)
        all_images.append(graphicID[0])    
print(len(all_images))
print(all_images)

all_images_with_initialized_count = []
for i in all_images:
    curtuple = (i, float(0.0))
    all_images_with_initialized_count.append(curtuple)

```

```python
file1 = "BC1.csv"
file2 = "BC2.csv"
file3 = "FC1.csv"
file4 = "FC2.csv"
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
df4 = pd.read_csv(file4)
```

```python
# binary-choice-bc1
all_images_with_initialized_count_bc1 = all_images_with_initialized_count.copy()
number_of_responses = df1.shape[0]
response_count_arr = []

for i in range(1, 129):
    
    chosen_row = df1["Q{x}_chosen".format(x = i)]
    val_count_dict = dict(chosen_row.value_counts(normalize =True))
    zippedlist = list(zip(val_count_dict.keys(), val_count_dict.values()))
    response_count_arr.extend(zippedlist)

for curtuple in response_count_arr:
    img_name = curtuple[0]
    img_count = curtuple[1]
    for i in range(0, len(all_images_with_initialized_count_bc1)):
        cur_initial_tuple = all_images_with_initialized_count_bc1[i]
        cur_initial_tuple_name = cur_initial_tuple[0]
        if img_name == cur_initial_tuple_name:
            all_images_with_initialized_count_bc1[i] = curtuple 
    
bc1_all_images_with_final_count = all_images_with_initialized_count_bc1.copy()
bc1_all_images_with_final_count.sort()
bc1_all_images_with_final_count[46]
```

```python
bc1_all_images_with_final_scores = []
bc2_all_images_with_final_scores = []
fc1_all_images_with_final_scores = []
fc2_all_images_with_final_scores = []

for i in bc1_all_images_with_final_count:
    bc1_all_images_with_final_scores.append(i[1])
for i in bc2_all_images_with_final_count:
    bc2_all_images_with_final_scores.append(i[1])
for i in fc1_all_images_with_final_count:
    fc1_all_images_with_final_scores.append(i[1])
for i in fc2_all_images_with_final_count:
    fc2_all_images_with_final_scores.append(i[1])
    
print(fc2_all_images_with_final_scores)
print(len(fc2_all_images_with_final_scores))
x_axis = []
for i in range(1, 257):
    x_axis.append(i)
print(x_axis)
```

```python
plt.scatter(x_axis, bc1_all_images_with_final_scores, label = "BC1")
plt.scatter(x_axis, bc2_all_images_with_final_scores, label = "BC2")
plt.title("Two-choice survey scatterplot")
plt.legend(loc = "upper right")
plt.xlabel("Image ID")
plt.ylabel("Probability that it was chosen in its image group")
plt.rcParams["figure.figsize"] = (8,7)
plt.show()

correlation, p_value = pearsonr(bc1_all_images_with_final_scores, bc2_all_images_with_final_scores)
print('Pearson\'s correlation: %.3f' % correlation)
```
